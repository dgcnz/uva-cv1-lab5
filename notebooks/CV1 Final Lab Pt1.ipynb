## **2.2 Building Visual Vocabulary**

Here, we will obtain visual words by clustering feature descriptors, so each cluster center is a visual word. Take a subset of all training images (this subset should contain images from ALL categories), extract SIFT descriptors from all of these images, and run k-means clustering (you can use your favourite k-means implementation) on these SIFT descriptors to build your visual vocabulary. Then, take the rest of the training images to calculate a visual dictionary.

You can also use less images, say 100 from each class (exclusive from the previous subset) if your computational resources are limited (remember to analyze the effects). Pre-defined cluster numbers will be the size of your vocabulary. In this question, set its size to 1000.

####  **` Q2.2: Building Visual Vocabulary. (10-pts)`**
Create a visual vocabulary by using K-means clustering. Remember to display the results when the vocabulary subset is 30\%, 40\% and 50\% amount of the training images. The vocabulary size is fixed 1000 in this question.

**Hint 1:** Remember first to debug all the code with a small amount of input images and only when you are sure that code functions correctly run it for training over the larger dataset.

**Hint 2:** You can achieve K-means clustering using either *sklearn* package or *scipy* package.

**Hint 3:** Results of K-means clustering could be shown using a scatter plot. In this case, the high-dimensional SIFT descriptors need to be brought down to a lower (2D) dimension  (*PCA* from *sklearn.decomposition* is one method to do this). For easy visualization, plotting up to 10 clusters would suffice. Check out an example implementation [here](https://www.askpython.com/python/examples/plot-k-means-clusters-python). Note that you are free to follow this approach, or you can show the K-means results in any other appropriate method that drives home your point.

```python
vocab_size = 1000
subset_percentages = [0.3, 0.4, 0.5]
sift = cv2.xfeatures2d.SIFT_create()
visual_words_per_subset = []
for percentage in subset_percentages:
    total_images = len(training_data)
    num_images = int(total_images * percentage)
    sift_descriptors = []
    class_labels = []
    for i in range(num_images):
        image = training_data[i]
        class_labels.append(training_label[i])
        keypoints, descriptors = sift.detectAndCompute(image, None)
        if descriptors is not None:
            sift_descriptors.extend(descriptors)
    print("Unique classes present in this subset:", np.unique(class_labels))
    kmeans = KMeans(n_clusters=vocab_size, random_state=0)
    kmeans.fit(sift_descriptors)
    visual_words = kmeans.cluster_centers_
    visual_words_per_subset.append(visual_words)
```

```python
for i, visual_words in enumerate(visual_words_per_subset):
    pca = PCA(n_components=2)
    reduced_visual_words = pca.fit_transform(visual_words)
    # Initialize the class object
    kmeans = KMeans(n_clusters=10)

    # predict the labels of clusters.
    label = kmeans.fit_predict(reduced_visual_words)

    # Getting the Centroids
    centroids = kmeans.cluster_centers_
    u_labels = np.unique(label)

    for i in u_labels:
        plt.scatter(
            reduced_visual_words[label == i, 0], reduced_visual_words[label == i, 1], label=i
        )
    plt.scatter(centroids[:, 0], centroids[:, 1], s=80, color="k")
    plt.legend()
    plt.show()
```

## **2.3 Encoding Features Using Visual Vocabulary**

Once we have a visual vocabulary, we can represent each image as a collection of visual words. For this purpose, we need to extract feature descriptors (with SIFT) and then assign each descriptor to the closest visual word from the vocabulary.

```python
image = training_data[-1]
keypoints, descriptors = sift.detectAndCompute(image, None)
descriptors = descriptors.astype("double")
visual_words = kmeans.predict(descriptors)
counts, bins = np.histogram(visual_words, density=True)
plt.hist(bins[:-1], bins, weights=counts)
```

## **2.6 Evaluation**

To evaluate your system, you should take all the test images from all classes and rank them based on each binary classifier. In other words, you should classify each test image with each classifier and then sort them based on the classification score. As a result, you will have five lists of test images. Ideally, you would have images with airplanes on the top of your list which is created based on your airplane classifier, and images with cars on the top of your list which is created based on your car classifier, and so on.

In addition to the qualitative analysis, you should measure the performance of the system quantitatively with the Mean Average Precision (mAP) over all classes. The mAP for a single class c is defined as
$$

\frac{1}{m_c} \sum_{i=1}^{n} \frac{f_c(x_i)}{i}

$$

where $n$ is the number of images ($n=50\times 5=250$), $m$ is the number of images of class $c$  ($m_c=50$), $x_i$ is the $i^{th}$ image in the ranked list $X = \left \{ x_1, x_2, \dots, x_n  \right \}$, and finally, $f_c$ is a function which returns the number of images of class $c$ in the first $i$ images if $x_i$ is of class $c$, and 0 otherwise. To illustrate, if we want to retrieve $R$ and we get the following sequence: $[R, R, T, R, T, T, R, T]$, then $n = 8$, $m = 4$, and $AP(R, R, T, R, T, T, R) = \frac{1}{4} \left (  \frac{1}{1} + \frac{2}{2} + \frac{0}{3} + \frac{3}{4} + \frac{0}{5} + \frac{0}{6} + \frac{4}{7} + \frac{0}{8} \right )$.

```python
def train_model(vocab_size, subset_percentage):
    sift = cv2.xfeatures2d.SIFT_create()
    visual_words_per_subset = []
    total_images = len(training_data)
    num_images = int(total_images * subset_percentage)
    sift_descriptors = []
    class_labels = []
    for i in range(num_images):
        image = training_data[i]
        class_labels.append(training_label[i])
        keypoints, descriptors = sift.detectAndCompute(image, None)
        if descriptors is not None:
            sift_descriptors.extend(descriptors)
    print("Unique classes present in this subset:", np.unique(class_labels))
    kmeans = KMeans(n_clusters=vocab_size, random_state=0)
    kmeans.fit(sift_descriptors)
    visual_words = kmeans.cluster_centers_
    visual_words_per_subset.append(visual_words)

    percentage = 1 - subset_percentage
    total_images = len(training_data)
    num_images = int(total_images * percentage)
    sift_descriptors = []
    class_wise_visual_word_stack = [[], [], [], [], []]

    for i in range(num_images):
        image = training_data[len(training_data) - 1 - i]
        keypoints, descriptors = sift.detectAndCompute(image, None)
        if descriptors is not None:
            descriptors = descriptors.astype("double")
            visual_words = kmeans.predict(descriptors)
            unique_elements, counts = np.unique(visual_words, return_counts=True)

            # Create a dictionary from the unique elements and their counts
            element_counts = dict(zip(unique_elements, counts))
            histogram_freq = np.zeros(vocab_size)
            np.put(histogram_freq, list(element_counts.keys()), list(element_counts.values()))
            class_wise_visual_word_stack[training_label[i]].append(histogram_freq)

    # class 1
    class_1 = np.array(class_wise_visual_word_stack[0])
    random_integers = np.random.randint(0, class_1.shape[0], 50)
    class_1_sample = class_1[random_integers]

    # class 2
    class_2 = np.array(class_wise_visual_word_stack[1])
    random_integers = np.random.randint(0, class_2.shape[0], 50)
    class_2_sample = class_2[random_integers]

    # class 3
    class_3 = np.array(class_wise_visual_word_stack[2])
    random_integers = np.random.randint(0, class_3.shape[0], 50)
    class_3_sample = class_3[random_integers]

    # class 4
    class_4 = np.array(class_wise_visual_word_stack[3])
    random_integers = np.random.randint(0, class_4.shape[0], 50)
    class_4_sample = class_4[random_integers]

    # class 5
    class_5 = np.array(class_wise_visual_word_stack[4])
    random_integers = np.random.randint(0, class_5.shape[0], 50)
    class_5_sample = class_5[random_integers]

    # class 1 SVM
    classifier_1 = svm.SVC(kernel="rbf", C=100.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="scale")
    train_negative_examples = np.vstack(
        (class_2_sample, class_3_sample, class_4_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_1.shape[0], 200)
    train_positive_examples = class_1[random_integers]
    classifier_1.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 2 SVM
    classifier_2 = svm.SVC(kernel="rbf", C=100.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="scale")
    train_negative_examples = np.vstack(
        (class_1_sample, class_3_sample, class_4_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_2.shape[0], 200)
    train_positive_examples = class_2[random_integers]
    classifier_2.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 3 SVM
    classifier_3 = svm.SVC(kernel="rbf", C=100.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="scale")
    train_negative_examples = np.vstack(
        (class_1_sample, class_2_sample, class_4_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_3.shape[0], 200)
    train_positive_examples = class_3[random_integers]
    classifier_3.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 4 SVM
    classifier_4 = svm.SVC(kernel="rbf", C=100.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="scale")
    train_negative_examples = np.vstack(
        (class_1_sample, class_3_sample, class_2_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_4.shape[0], 200)
    train_positive_examples = class_4[random_integers]
    classifier_4.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 5 SVM
    classifier_5 = svm.SVC(kernel="rbf", C=100.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="scale")
    train_negative_examples = np.vstack(
        (class_1_sample, class_2_sample, class_3_sample, class_4_sample)
    )
    random_integers = np.random.randint(0, class_5.shape[0], 200)
    train_positive_examples = class_5[random_integers]
    classifier_5.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    return classifier_1, classifier_2, classifier_3, classifier_4, classifier_5, kmeans
```

```python
def train_model_svm_tune(vocab_size, subset_percentage):
    sift = cv2.xfeatures2d.SIFT_create()
    visual_words_per_subset = []
    total_images = len(training_data)
    num_images = int(total_images * subset_percentage)
    sift_descriptors = []
    class_labels = []
    for i in range(num_images):
        image = training_data[i]
        class_labels.append(training_label[i])
        keypoints, descriptors = sift.detectAndCompute(image, None)
        if descriptors is not None:
            sift_descriptors.extend(descriptors)
    print("Unique classes present in this subset:", np.unique(class_labels))
    kmeans = KMeans(n_clusters=vocab_size, random_state=0)
    kmeans.fit(sift_descriptors)
    visual_words = kmeans.cluster_centers_
    visual_words_per_subset.append(visual_words)

    percentage = 1 - subset_percentage
    total_images = len(training_data)
    num_images = int(total_images * percentage)
    sift_descriptors = []
    class_wise_visual_word_stack = [[], [], [], [], []]

    for i in range(num_images):
        image = training_data[len(training_data) - 1 - i]
        keypoints, descriptors = sift.detectAndCompute(image, None)
        if descriptors is not None:
            descriptors = descriptors.astype("double")
            visual_words = kmeans.predict(descriptors)
            unique_elements, counts = np.unique(visual_words, return_counts=True)

            # Create a dictionary from the unique elements and their counts
            element_counts = dict(zip(unique_elements, counts))
            histogram_freq = np.zeros(vocab_size)
            np.put(histogram_freq, list(element_counts.keys()), list(element_counts.values()))
            class_wise_visual_word_stack[training_label[i]].append(histogram_freq)

    # class 1
    class_1 = np.array(class_wise_visual_word_stack[0])
    random_integers = np.random.randint(0, class_1.shape[0], 50)
    class_1_sample = class_1[random_integers]

    # class 2
    class_2 = np.array(class_wise_visual_word_stack[1])
    random_integers = np.random.randint(0, class_2.shape[0], 50)
    class_2_sample = class_2[random_integers]

    # class 3
    class_3 = np.array(class_wise_visual_word_stack[2])
    random_integers = np.random.randint(0, class_3.shape[0], 50)
    class_3_sample = class_3[random_integers]

    # class 4
    class_4 = np.array(class_wise_visual_word_stack[3])
    random_integers = np.random.randint(0, class_4.shape[0], 50)
    class_4_sample = class_4[random_integers]

    # class 5
    class_5 = np.array(class_wise_visual_word_stack[4])
    random_integers = np.random.randint(0, class_5.shape[0], 50)
    class_5_sample = class_5[random_integers]

    # class 1 SVM
    classifier_1 = svm.SVC(kernel="rbf", C=50.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="auto")
    train_negative_examples = np.vstack(
        (class_2_sample, class_3_sample, class_4_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_1.shape[0], 200)
    train_positive_examples = class_1[random_integers]
    classifier_1.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 2 SVM
    classifier_2 = svm.SVC(kernel="rbf", C=50.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="auto")
    train_negative_examples = np.vstack(
        (class_1_sample, class_3_sample, class_4_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_2.shape[0], 200)
    train_positive_examples = class_2[random_integers]
    classifier_2.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 3 SVM
    classifier_3 = svm.SVC(kernel="rbf", C=50.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="auto")
    train_negative_examples = np.vstack(
        (class_1_sample, class_2_sample, class_4_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_3.shape[0], 200)
    train_positive_examples = class_3[random_integers]
    classifier_3.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 4 SVM
    classifier_4 = svm.SVC(kernel="rbf", C=50.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="auto")
    train_negative_examples = np.vstack(
        (class_1_sample, class_3_sample, class_2_sample, class_5_sample)
    )
    random_integers = np.random.randint(0, class_4.shape[0], 200)
    train_positive_examples = class_4[random_integers]
    classifier_4.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    # class 5 SVM
    classifier_5 = svm.SVC(kernel="rbf", C=50.0, probability=True, 𝑔𝑎𝑚𝑚𝑎="auto")
    train_negative_examples = np.vstack(
        (class_1_sample, class_2_sample, class_3_sample, class_4_sample)
    )
    random_integers = np.random.randint(0, class_5.shape[0], 200)
    train_positive_examples = class_5[random_integers]
    classifier_5.fit(
        np.vstack((train_positive_examples, train_negative_examples)),
        np.hstack((np.ones(len(train_positive_examples)), np.zeros(len(train_negative_examples)))),
    )

    return classifier_1, classifier_2, classifier_3, classifier_4, classifier_5, kmeans
```
